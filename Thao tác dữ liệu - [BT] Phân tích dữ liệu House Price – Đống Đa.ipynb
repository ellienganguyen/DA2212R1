```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\DA-Class\DA2212R1\End_of_term\ApartmentTrading.csv")
df.info()
df.columns
#df.shape
```

```python
#2. Lọc ra các bản ghi bán nhà riêng tại phường Trung liệt hoặc phường Khâm Thiên ===> ko có trong data ==> lẤY Data KHÁC

tmp = df[(df['ten_duong']=='Phố Liễu Giai') | (df['ten_phuong']== 'Phường Khâm Thiên')]
tmp

# df_filtered = df.loc[((df['ten_duong'].str.contains('Phố Liễu Giai')) | (df['ten_phuong'].str.contains('Phường Khâm Thiên')))]
# print(df_filtered)

```

```python

#3. Lọc các thông tin Địa chỉ, Giá, Hướng nhà, Hướng ban công của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên.

df2 = df[(df['phong_ngu'] >=3) & ((df['so_do'] == "Sổ đỏ") | (df['so_do'] == "Đã có sổ") | (df['so_do'] == "Sổ hồng/Sổ đỏ"))].filter(['dien_tich', 'gia', 'phong_ngu', 'huong_nha', 'huong_ban_cong','so_do', 'gia_m2'])
print(df2)
```

```python
#4. Với mỗi loại nhà đất-> mõi quận, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.

df3 = df.groupby('ten_quan')['gia'].agg(['max','min','mean'])
df3
```

```python
#5. Tính trung bình cộng số phòng ngủ, số phòng vệ sinh, số tầng của mỗi phường

df4 = df.groupby('ten_phuong')['phong_ngu', 'so_tang'].agg('mean')
print(df4)
```
