```python
# 2.Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; 
#nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; 
#nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9; 
#nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;
tmp = df.copy()
tmp['Month']=pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M').dt.month
tmp['Year']=pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M').dt.year
tmp
```

```python
conditions = [(tmp['Month'] == 1) | (tmp['Month'] == 2) | (tmp['Month'] == 3),
              (tmp['Month'] == 4) | (tmp['Month'] == 5) | (tmp['Month'] == 6),
              (tmp['Month'] == 7) | (tmp['Month'] == 8) | (tmp['Month'] == 9),
              (tmp['Month'] == 10) | (tmp['Month'] == 11) | (tmp['Month'] == 12)]
choices = ['1', '2', '3', '4']
tmp['Quarter'] = np.select(conditions, choices, default=' ')
tmp 
```

```python
#3. Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice

tmp['Amount'] = tmp['Quantity'] * tmp['UnitPrice']
tmp
```

```python
#4. Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu là ‘France’ ngược lại là 0%.

conditions = [(tmp['Country'] == 'United Kingdom') & (tmp['Quarter'] == 4),
              (tmp['Country'] == 'France')]
choices = ['10%', '5%']
tmp['Discount'] = np.select(conditions, choices, default='0%')
tmp 
```

```python
# 5. Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.

# tmp['Discount'] = tmp['Discount'].replace('[^0-9\.-]', '', regex=True) #Loại bỏ các ký tự không phải số trong cột Discount bằng phương thức replace 
# tmp['Discount'] = tmp['Discount'].astype(float)
# tmp['Discount'] = tmp['Discount'] / 100

tmp['Total'] = tmp['Amount'] -tmp['Amount']*tmp['Discount']
tmp

```
