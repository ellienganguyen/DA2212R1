Bộ dữ liệu FoodPrice_in_Turkey chứa dữ liệu chứa dữ liệu về giá cả các mặt hàng thực phẩm tại Turkey, các thông tin được thống kê bao gồm:
- Place: Nơi bán
- ProductID: Mã sản phẩm
- ProductName: Tên sản phẩm
- UmId: Mã đơn vị đo lường
- UmName: Tên đơn vị đo lường
- Month: Tháng
- Year: Năm
- Price: Giá

```python
import pandas as pd
import numpy as np

data = pd.read_csv('Datasets\FoodPrice_in_Turkey.csv', encoding = "ISO-8859-1")
data.head()
```

```python
# Bộ dữ liệu chứa bao nhiêu dòng, bao nhiêu cột
print(data.info())
```

```python
#Thang đo tương ứng của các thuộc tính được lưu trữ
#Place, ProcutName, UmName: thuộc tính định tính - norminal
#ProductId: thuộc tính định tính - norminal
#Month, Year: thuộc tính định lượng - interval
#UmId, Price: thuộc tính định lượng - ratio
```

```python
#Giá trung bình của từng loại thực phẩm là bao nhiêu
AVR_price = data.groupby('ProductName')['Price'].mean()
AVR_price
```

```python
#Bạn đề xuất tất cả các yêu cầu phân tích, hiện tại bạn chưa làm được, 
#nhưng hãy chúng ta sẽ giải quyết ở các bài học sau

# Vẽ biểu đồ phân bố giá của các loại sản phẩm
print ("trung bình giá bán: " + str(data.Price.mean()))
print ("trung vị của giá bán: "+ str(data.Price.median()))
import matplotlib.pyplot as plt

plt.hist(data.Price, bins = 200)
plt.title("Phân bố giá bán các sản phẩm")
plt.xlabel("Giá bán")
plt.ylabel("Số sản phẩm")
```
